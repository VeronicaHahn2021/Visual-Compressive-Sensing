```python
import os
import sys
sys.path.append('../../')
from src.compress_sensing_library import reconstruct, generate_V1_variables, error_calculation
from matplotlib import pyplot as plt
from PIL import Image, ImageOps
from sklearn.linear_model import Lasso
import numpy as np
import pywt
from pywt import wavedecn
```

# Signal Extension
### zero: zero-padding
### constant: constant-padding
### symmetric
### periodic
### smooth

```python
image_nm_list = ['image/city_part2.png', 'image/city_part3.png', 'image/tree_part1.jpg']
for image_nm in image_nm_list :
    save_nm = image_nm.split('/')[1].split('.')[0]
    img = Image.open(os.path.join('../../' + image_nm))
    img = ImageOps.grayscale(img)
    img_arr = np.asarray(img)
    plt.axis('off')
    plt.title('Original city image')
    plt.imshow(img_arr, 'gray')

    num_cell = 500
    cell_size = 7
    sparse_freq = 2
    alpha = 1
    w =  'db2'
    lv = 4


    n, m = img_arr.shape
    samp = np.floor(n * m * .2).astype(int)
    rand_index = np.random.randint(0, n * m, samp)
    mode_list = ['zero', 'constant', 'symmetric', 'periodic', 'smooth']
    C3D, y = generate_V1_variables(num_cell, cell_size, sparse_freq, img_arr)
    for mode in mode_list :
        dwt_sample = wavedecn(C3D[0], wavelet = w, mode = mode, level = lv, axes = [0, 1], )
        coeff, coeff_slices, coeff_shapes = pywt.ravel_coeffs(dwt_sample)
        theta = np.zeros((len(C3D), len(coeff)))
        theta[0, :] = coeff 

        # Loop the wavedecn to fill theta
        for i in range(samp):
            theta_i = wavedecn(C3D[i], wavelet= w, level = lv, mode = mode, axes = [0, 1])
            theta[i, :] = pywt.ravel_coeffs(theta_i)[0]

        mini = Lasso(alpha = alpha, fit_intercept = False)
        mini.fit(theta, y)

        s = mini.coef_

        s_unravel = pywt.unravel_coeffs(s, coeff_slices, coeff_shapes)
        reconst = pywt.waverecn(s_unravel, w, mode = mode)
        plt.imshow(reconst, 'gray', vmin = 0, vmax = 255)

        plt.axis('off')
        title = 'dwt_V1 {img_nm} reconst mode {mode}'.format(img_nm = save_nm,mode = mode)
        plt.title(title)
        plt.savefig(os.path.join("../../", "figures/signal_extension/{title}.jpg".format(title = title.replace(" ", "_"))), dpi = 300)
        plt.show()
```
