# Classical

```python
a = np.arange(0, 30, 1)
x_temp = np.random.permutation(a)[:20]
y_temp = np.random.permutation(a)[:20]

fig = plt.figure(figsize = (13, 4))
plt.subplots_adjust(top=0.6)
plt.title('Classical weights (random Index: 500)')
plt.axis('off')
for i in range(20):
    ax = fig.add_subplot(2, 10, i+1)
    plt.scatter(x_temp[i], y_temp[i], color = 'red', marker = '.')
    plt.xlim([0, 30])
    plt.ylim([0, 30])
    
    axes = plt.gca()
    axes.get_xaxis().set_visible(False)
    axes.get_yaxis().set_visible(False)
fig.savefig("result/supplement_figure/classical_weights_figure.png", dpi = 300)
```

# V1

```python
num_cell = 500
dim = (30, 30)
cell_size = 7
sparse_freq = 2

W = V1_weights(num_cell, dim, cell_size, sparse_freq)
```

```python
fig = plt.figure(figsize = (13, 4))
plt.subplots_adjust(top=0.6)
plt.title('Random V1 weights (num cell: 500, cell size: 7, sparse freq: 2)')
plt.axis('off')
for i in range(20):
    ax = fig.add_subplot(2, 10, i+1)
    cell_rf = W[i]
    cell_rf = cell_rf.reshape(dim)
    plt.imshow(cell_rf, cmap="RdBu")
    plt.clim([-3, 3])
    axes = plt.gca()
    axes.get_xaxis().set_visible(False)
    axes.get_yaxis().set_visible(False)
#     plt.axis('off')
fig.savefig("result/supplement_figure/V1_weights_figure.png", dpi = 300)
```
