```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as la

from src.structured_random_features.src.models.weights import V1_weights

# Packages for fft and fitting data
from scipy import fftpack as fft
from sklearn.linear_model import Lasso

# Package for importing image representation
from PIL import Image, ImageOps

from src.V1_Compress import generate_Y, compress
import pandas as pd
from dask.distributed import Client, progress
import seaborn as sns
import time
import os.path
```

```python
image_path = 'image/city_part3.png'
num_cell = 500
cell_size = 7
sparse_freq = 2
alpha = 0.01

filter_dim = (30, 30)
filt = np.zeros(filter_dim)

img = Image.open(image_path)
img = ImageOps.grayscale(img)
img_arr = np.asarray(img)
dim = img_arr.shape


img_arr_pt1 = img_arr[:15, :15]
img_arr_pt2 = img_arr[:15, 15:30]
img_arr_pt3 = img_arr[15:, :15]
img_arr_pt4 = img_arr[15:, 15:30]
img_pt = [img_arr_pt1, img_arr_pt2, img_arr_pt3, img_arr_pt4]
dim_pt = img_arr_pt1.shape
n_pt = dim_pt[0]
m_pt = dim_pt[1]

res = np.zeros(dim)
count = 0

fig, axs = plt.subplots(2, 4)
for i, pt in enumerate(img_pt):
    axs[0, count].imshow(pt)
    axs[0, count].set_title("real pt {n}".format(n = count))
#     plt.show()
    W = V1_weights(num_cell, dim_pt, cell_size, sparse_freq) 
    y = generate_Y(W, pt)
    W_model = W.reshape(num_cell, n_pt, m_pt)
    theta, reform, s = compress(W_model, y, alpha)
    axs[1, count].imshow(reform)
    axs[0, count].set_title("reconstructed pt {n}".format(n = count))
    error_h = np.linalg.norm(pt - reform, 'fro') / np.sqrt(m_pt*n_pt)
#     print("Error:", error_h)
    if (i == 0):
        res[:15, :15] = reform
    elif (i == 1):
        res[:15, 15:30] = reform
    elif (i == 2):
        res[15:, :15] = reform
    elif (i == 3):
        res[15:, 15:30] = reform
    count+=1
plt.imshow(img_arr - res)
plt.show()
plt.imshow(reform)
```
