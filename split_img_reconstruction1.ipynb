```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as la

from src.structured_random_features.src.models.weights import V1_weights

# Packages for fft and fitting data
from scipy import fftpack as fft
from sklearn.linear_model import Lasso

# Package for importing image representation
from PIL import Image, ImageOps

from src.V1_Compress import generate_Y, compress
import pandas as pd
from dask.distributed import Client, progress
import seaborn as sns
import time
import os.path
```

```python
image_path = 'image/city_part3.png'
num_cell_pt = 125
num_cell = 500
cell_size = 7
sparse_freq = 2
alpha = 0.01

filter_dim = (30, 30)
filt = np.zeros(filter_dim)

img = Image.open(image_path)
img = ImageOps.grayscale(img)
img_arr = np.asarray(img)
dim = img_arr.shape
n, m = dim

img_arr_pt1 = img_arr[:15, :15]
img_arr_pt2 = img_arr[:15, 15:30]
img_arr_pt3 = img_arr[15:, :15]
img_arr_pt4 = img_arr[15:, 15:30]
img_pt = [img_arr_pt1, img_arr_pt2, img_arr_pt3, img_arr_pt4]
dim_pt = img_arr_pt1.shape
n_pt = dim_pt[0]
m_pt = dim_pt[1]

res = np.zeros(dim)
count = 0

fig, axs = plt.subplots(2, 4)
plt.figure(figsize=(10, 10))
W_pt = V1_weights(num_cell_pt, dim_pt, cell_size, sparse_freq)
W_whole = V1_weights(num_cell, dim, cell_size, sparse_freq)
for i, pt in enumerate(img_pt):
    axs[0, count].imshow(pt)
    axs[0, count].set_title("real pt {n}".format(n = count))
#     axs[0, count].show()
    
    y_pt = generate_Y(W_pt, pt)
    W_model_pt = W_pt.reshape(num_cell_pt, n_pt, m_pt)
    theta_pt, reform_pt, s_pt = compress(W_model_pt, y_pt, alpha)
    axs[1, count].imshow(reform_pt)
    axs[1, count].set_title("reconstructed pt {n}".format(n = count))
#     axs[1, count].show()
    error_h = np.linalg.norm(pt - reform_pt, 'fro') / np.sqrt(m_pt*n_pt)
#     print("Error:", error_h)
    if (i == 0):
        res[:15, :15] = reform_pt
    elif (i == 1):
        res[:15, 15:30] = reform_pt
    elif (i == 2):
        res[15:, :15] = reform_pt
    elif (i == 3):
        res[15:, 15:30] = reform_pt
    count+=1
    
y = generate_Y(W_whole, img_arr)
W_model = W_whole.reshape(num_cell, n, m)
theta_whole, reform_whole, s_whole = compress(W_model, y, alpha)
# plt.imshow(img_arr - res)
# plt.show()
# plt.imshow(reform)
```

```python
plt.imshow(img_arr)
plt.title("original image")
plt.show()

plt.imshow(res.astype(int))
plt.title("reconstructed splited image")
plt.show()

plt.imshow(img_arr - res)
plt.title("error image for split reconstruction")
plt.show()

plt.imshow(reform_whole.astype(int))
plt.title("reconstructed whole image")
plt.show()
plt.imshow(img_arr - reform_whole)
plt.title("error image for whole reconstruction")
plt.show()

print("error rate for split image is {a}, for whole image is {b}".format(
    a = np.linalg.norm(img_arr - res, 'fro') / np.sqrt(m_pt*n_pt),
    b = np.linalg.norm(img_arr - reform_whole, 'fro') / np.sqrt(m*n)))
```
